```python
# Imports
import pickle
import community as community_louvain  
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
# Load the graph
with open("road_graph.pkl", "rb") as f:
    G_main = pickle.load(f)

# Extract Largest Weakly Connected Component
largest_cc = max(nx.weakly_connected_components(G_main), key=len)
G_largest = G_main.subgraph(largest_cc).copy()

print(f"Nodes: {G_largest.number_of_nodes()}, Edges: {G_largest.number_of_edges()}")
```

```python
# Convert to undirected
G_largest_undirected = G_largest.to_undirected()

# Compute Louvain communities
partition = community_louvain.best_partition(G_largest_undirected)

# Save to CSV
pd.DataFrame(list(partition.items()), columns=["node", "community"]).to_csv(
    "output/communities.csv", index=False
)

print("✅ Louvain communities saved to output/communities.csv")
```

```python
# Take a sample of nodes (e.g. 500 random ones)
sample_nodes = random.sample(list(G_largest.nodes()), min(500, len(G_largest)))
G_sample = G_largest.subgraph(sample_nodes).copy()

# Layout (faster than spring_layout)
pos = nx.kamada_kawai_layout(G_sample)

# Draw without labels (much faster)
plt.figure(figsize=(10, 10))
nx.draw_networkx_nodes(G_sample, pos, node_size=30, 
                       node_color=[partition[n] for n in G_sample.nodes()], cmap=plt.cm.tab20)
nx.draw_networkx_edges(G_sample, pos, alpha=0.3)
plt.axis("off")
plt.show()
```

```python
# ---- Community Analysis ----

# Convert partition dict → DataFrame
df_com = pd.DataFrame(list(partition.items()), columns=["node", "community"])

# Community counts and sizes
print("Number of communities:", df_com["community"].nunique())
print("\nTop 10 biggest communities:")
print(df_com["community"].value_counts().head(10))

# Add PageRank scores
pr = nx.pagerank(G_largest)
df_com["pagerank"] = df_com["node"].map(pr)

# Top 5 nodes per community by PageRank
df_top = df_com.sort_values(["community", "pagerank"], ascending=[True, False])
top_nodes = df_top.groupby("community").head(5)
print("\nTop 5 nodes per community:")
print(top_nodes.head(20))  # preview first few communities

# Save results
df_com.to_csv("output/community_analysis.csv", index=False)
print("\n✅ Saved community analysis to output/community_analysis.csv")
```

```python
# Load the graph with pickle
with open("hanoi_directed_graph.pkl", "rb") as f:
    G_main = pickle.load(f)

print(f"Graph has {G_main.number_of_nodes()} nodes and {G_main.number_of_edges()} edges")
print("Directed:", G_main.is_directed())

# Largest Weakly Connected Component
largest_wcc = max(nx.weakly_connected_components(G_main), key=len)
G_largest = G_main.subgraph(largest_wcc).copy()

print(f"Largest WCC size: {len(G_largest)}")

# Plot the largest weakly connected component
nodes_sample = random.sample(list(G_largest.nodes()), 500)  # pick 500 nodes
G_sample = G_largest.subgraph(nodes_sample)

pos = nx.spring_layout(G_sample, k=0.1, iterations=30)
nx.draw(G_sample, pos, node_size=20, edge_color="gray", node_color="blue", with_labels=False)
plt.title("Sample of Largest WCC")
plt.show()
```
